**Changing Column names of data1 frame**

```python
data1.rename(columns={data1.columns[0]:'trip_id', data1.columns[1]:'start_time', data1.columns[2]:'end_time', data1.columns[3]:'bikeid', data1.columns[4]: 'tripduration', data1.columns[5]: 'from_station_id', data1.columns[6]: 'from_station_name', data1.columns[7]: 'to_station_id', data1.columns[8]: 'to_station_name', data1.columns[9]: 'usertype',data1.columns[10]: 'gender',data1.columns[11]: 'birthyear'},inplace=True)

data1.head()
```

**Making start_time and end_time columns consistent in data2 Dataframe**

```python
# Convert "start_time" and "end_time" columns to datetime
data2['start_time'] = pd.to_datetime(data2['start_time'])
data2['end_time'] = pd.to_datetime(data2['end_time'])

# Format "start_time" and "end_time" columns with HH:MM:SS
data2['start_time'] = data2['start_time'].dt.strftime('%Y-%m-%d %H:%M:%S')
data2['end_time'] = data2['end_time'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Display the resulting DataFrame with "start_time" and "end_time" in HH:MM:SS format
print(data2[['start_time', 'end_time']])
```

**CONCATENATE ALL 4 FILES**

```python
data = pd.concat([data1, data2, data3, data4], axis=0, ignore_index=True)
```

```python
data.shape
```

```python
data.head()
```

**Dropping Columns Gender and Birthyear**

```python
data.drop(['gender','birthyear'],axis=1,inplace=True)
data.head()
```

**Adding A New Column 'ride_length' and then converting it to int64**

```python
# Convert "start_time" and "end_time" columns to datetime
data['start_time'] = pd.to_datetime(data['start_time'])
data['end_time'] = pd.to_datetime(data['end_time'])

data['ride_length'] = data['end_time'] - data['start_time']

# Convert ride_length to integer representing seconds
data['ride_length_seconds'] = data['ride_length'].dt.total_seconds().astype('int64')
```

```python
data.head(10)
```

**Add A column for 'day_of_week'**

```python
data['start_time'] = pd.to_datetime(data['start_time'])

# Calculate the day of the week where 0 = Monday and 6 = Sunday
data['day_of_week'] = data['start_time'].dt.dayofweek

# Adjust the day numbering to match the requested format (1 = Sunday, 7 = Saturday)
data['day_of_week'] = (data['day_of_week'] + 1) % 7 + 1

# Display the resulting DataFrame with the "day_of_week" column
print(data[['start_time', 'day_of_week']])
```

**Convert start_time and end_time columns to datetime64 datatype and ride_length datatype to timedelta64**

```python
# Convert "start_time" and "end_time" columns to datetime
data['start_time'] = pd.to_datetime(data['start_time'])
data['end_time'] = pd.to_datetime(data['end_time'])

# Convert "ride_length" column to timedelta
data['ride_length'] = pd.to_timedelta(data['ride_length'])
```

```python
data.drop(['ride_length'],axis=1,inplace=True)
```

**Adding columns for Quarter, Month and Week by Parsing the start_time column**

```python
data['start_week'] = data['start_time'].dt.week
data['start_month'] = data['start_time'].dt.month
data['start_quarter'] = data['start_time'].dt.quarter

```
